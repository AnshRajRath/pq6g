```python
#casar cipher encryption
i = str(input("enter the string:"))
x = int(input("enter the shift number:"))
def cc(i,x):
    r=""
    for _ in i:
        if _.isalpha():
            if _.islower():
                r+=chr((ord(_)-ord('a') + x)%26 + ord('a'))
            elif _.isupper():
                r+=chr((ord(_)-ord('A') + x)%26 + ord('A'))
        else:
            r+=_
    return r
print(cc(i,x))
```

```python
#casar cipher decryption
i = str(input("enter the string:"))
x = int(input("enter the shift number:"))
def ccd(i,x):
    r=""
    for _ in i:
        if _.isalpha():
            if _.islower():
                r+=chr((ord(_)-ord('a')- x)%26 + ord('a'))
            elif _.isupper():
                r+=chr((ord(_)-ord('A') - x)%26 + ord('A'))
        else:
            r+=_
    return r
print(ccd(i,x))
```

```python
#vignere cipher encryption
i=str(input("enter the string:"))
k=str(input("enter the key string:"))
def vce(i,k):
    r=""
    ki = 0
    for _ in i:
        if _.isalpha():
            shift = ord(k[ki%len(k)].lower()) - ord('a')
            if _.islower():
                r+=chr(((ord(_)-ord('a')) + shift)%26 + ord('a'))
            elif _.isupper():
                r+=chr(((ord(_)-ord('A')) + shift)%26 + ord('A'))
            ki += 1
        else:
            r+=_
    return r
print(vce(i,k)) 
```

```python
#vignere cipher decryption
i=str(input("enter the string:"))
k=str(input("enter the key string:"))
def vcd(i,k):
    r=""
    ki = 0
    for _ in i:
        if _.isalpha():
            shift = ord(k[ki%len(k)].lower()) - ord('a')
            if _.islower():
                r+=chr(((ord(_)-ord('a')) - shift)%26 + ord('a'))
            elif _.isupper():
                r+=chr(((ord(_)-ord('A')) - shift)%26 + ord('A'))
            ki += 1
        else:
            r+=_
    return r
print(vcd(i,k))
```
